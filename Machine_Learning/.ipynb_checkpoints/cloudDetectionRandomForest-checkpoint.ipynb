```python
import numpy as np
# Grab Dataset
data = np.genfromtxt('../data.csv', delimiter=',')
```

```python
# Split data into y (label) and x (features)
y = data[:,0]
x = data[:,1:]
x = np.nan_to_num(x, nan=0.0)
```

```python
# Any multi-layer cloud labels set to 1, turn into classification [cloud, no cloud]=[1,0]
y = np.where(y >= 1, 1, y)
print(y[0])
```

```python
from sklearn.model_selection import train_test_split
# Split data into Train/Test/Validation 80/20
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
# Scale data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

2. Build Random Forest
3. Train

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

rf_classifier = RandomForestClassifier(n_estimators=300, random_state=42)

rf_classifier.fit(X_train, y_train)

y_pred = rf_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

4. Gather Feature Importance

```python
import matplotlib.pyplot as plt

column_names = ['Band 20','Band 27','Band 28','Band 29','Band 30','Band 31','Band 32','Band 33','Band 34','Band 35','Band 36','Band 17','Band 18','Band 19','Band 26','Band  1','Band  2','Band  3','Band  4','Band  5','Band  6','Band  7']

# Get feature importances
feature_importances = rf_classifier.feature_importances_
sorted_indices = feature_importances.argsort()[::-1]
sorted_indices = sorted_indices[::-1]

# Print and plot feature importances
print("\nFeature Importances:")
plt.figure(figsize=(5, 6))
plt.barh(range(x.shape[1]), feature_importances[sorted_indices], color='maroon')
plt.yticks(range(x.shape[1]), [column_names[i] for i in sorted_indices], rotation=0, ha='right', fontname='Verdana')
plt.ylabel('Feature', fontname='Verdana')
plt.xlabel('F score', fontname='Verdana')
plt.title('Feature Importances', fontname='Verdana', fontsize=14)
plt.show()
```

## Downsample to compare to QSVM

```python
# Generate 10000 random indices
np.random.seed(7)
random_indices = np.random.choice(len(x), size=10000, replace=False)

# Select samples using the random indices
x = x[random_indices,:]
y = y[random_indices]
```

```python
# cut down on features
# 0 and 6 were chosen from Random Forest's Feature Importance
x = x[:,[0,6]]
```

```python
from sklearn.model_selection import train_test_split
# Split data into Train/Test/Validation 10/90
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.9, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
# Scale data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

rf_classifier = RandomForestClassifier(n_estimators=300, random_state=42)

rf_classifier.fit(X_train, y_train)

y_pred = rf_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
